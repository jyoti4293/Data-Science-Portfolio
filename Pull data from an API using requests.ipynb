#### Using YouTube API to get the data related to search and video statistics

Reference documets-
1. https://www.slickremix.com/docs/get-api-key-for-youtube/
2. https://developers.google.com/youtube/v3/docs

```python
#import libraries
import requests
import pandas
import time
```

```python
#get the API key to make API calls
API_Key=""
Channel_ID=""
pageToken=""
```

```python
#defining a function to get the video_statistics
def video_stats(video_id):
    
    #create url to fetch data after making an API call
    url_video_stats="https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+API_Key
    
    #store the data in json format
    response_video_stats=requests.get(url_video_stats).json()
    
    #parse the data and store in lists
    view_count=response_video_stats['items'][0]['statistics']['viewCount']
    like_count=response_video_stats['items'][0]['statistics']['likeCount']
    comment_count=response_video_stats['items'][0]['statistics']['commentCount']
    
    return view_count, like_count, comment_count
```

```python
#Another function to get the video_id and then use first function to get video_stats for those video_ids
def get_dataframe(df):
    
    #create url to fetch the data
    url="https://www.googleapis.com/youtube/v3/search?key="+API_Key+"&channelId="+Channel_ID+"&part=snippet,id&order=date&maxResults=10000"+pageToken
    
    #store the data in json format
    response=requests.get(url).json()

    #wait till the data is fetched
    time.sleep(1)
    
    #parse the data for each observation and store in lists
    for video in response['items']:
    
        if video['id']['kind']=="youtube#video":
        
            video_id=video['id']['videoId']
            video_title=video['snippet']['title']
            upload_date=video['snippet']['publishedAt']
            upload_date=upload_date.split("T")[0]
            
            view_count, like_count, comment_count=video_stats(video_id)
            
            #save the data into a dataframe
            df=df.append({'video_id':video_id,
                  'video_title':video_title,
                  'upload_date':upload_date,
                  'view-count':view_count,
                  'like_count':like_count,
                  'comment_count':comment_count},ignore_index=True)
            
    return df
        
```
